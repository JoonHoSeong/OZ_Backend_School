### 1. 변수명 게임에 들어 있는 값 중 '홀'만 출력되도록 코드를 작성해 주세요

     게임 = “홀짝홀짝홀짝”

     출력 예시 : “홀홀홀”

<aside>
🔑 정답 : **print(게임[::2])**

문자열 슬라이싱을 이용하면 게임 변수에서 홀수 인덱스의 문자만 추출할 수 있습니다. 파이썬의 인덱스는 0부터 시작하므로, '홀' 문자는 0, 2, 4 등의 인덱스에 위치합니다.

따라서 '홀' 문자만 출력하려면 다음과 같이 코드를 작성할 수 있습니다.

```python
print(게임[::2])
```

위 코드에서 `[::2]`는 게임 변수의 0번 인덱스부터 시작해서 2의 간격으로 문자를 선택하라는 의미입니다. 이렇게 하면 '홀' 문자만 선택되어 출력됩니다.

</aside>

### 2. 변수명 전화번호에 들어 있는 번호를 모두 붙여서 출력되도록 코드를 작성해주세요

    전화번호 = “010-1234-5678

    출력 예시 : “01012345678”

<aside>
🔑 정답 : **‘’.join(전화번호.split(’-’))**

`split()` 함수를 이용하여 하이픈('-')을 제거할 수도 있습니다. `split()` 함수는 문자열을 특정 구분자를 기준으로 나누어 리스트로 반환합니다.

따라서 전화번호에서 하이픈을 제거하려면 먼저 `split()` 함수로 전화번호를 하이픈을 기준으로 나눈 후, `join()` 함수를 이용해 다시 합칠 수 있습니다.

다음은 해당 코드입니다:

```python
print(‘’.join(전화번호.split(’-’)))
```

위 코드에서 `전화번호.split('-')`는 전화번호를 하이픈('-')을 기준으로 나누어 리스트를 반환하고, `''.join()`은 이 리스트의 요소들을 빈 문자열('')로 연결하여 하나의 문자열로 만듭니다. 

</aside>

### 3. 창의성, 시장성, 안정성, 수익성에 대한 각각의 평가 중 아래 기준에 부합하지 못하면 창업지원을 받을 수 없습니다.

- 수익성 60점 이상
- 창의성과 시장성 50점 이상
- 시장성과 수익성을 더한 값의 절반이 안정성 보다 높아야 함

다음 점수가 입력 되었을때 창업 지원이 가능한지 여부를 True 또는 False로 출력해주세요

- 창의성 : 62 / 시장성 : 71 / 안정성 : 68 / 수익성 : 66

<aside>
🔑 정답 : **print(수익성 >= 60 and 창의성 >= 50 and 시장성 >=50 and ((시장성+수익성) / 2) > 안정성), 결과값 : True**

```python
창의성, 시장성, 안정성, 수익성 = 62, 71, 68, 66

print(수익성 >= 60 and 창의성 >= 50 and 시장성 >=50 and ((시장성+수익성) / 2) > 안정성)
```

1. 각 변수에 값을 할당합니다.
2. 창업 지원에 필요한 조건을 순서대로 코드로 작성합니다.
3. 각각의 조건은 모두 만족해야하므로 다음 조건을 연결할때는 AND 연산자를 사용합니다.
만일, 주어진 주건 중 하나만 만족하거나 두개만 만족해도 되는 경우라면 OR로 해결할 수 있습니다.
4. 비교 연산자의 결과는 True 또는 False로 나오므로 바로 출력합니다.
</aside>
