### 2. 문제

밑변과 높이가 주어졌을때 피타고라스 정의를 이용해 빗면을 구할 수 있습니다.

아래 주어진 밑변과 높이를 이용해 빗면을 구해주세요

image.png

<aside>
🔑 정답 : **((9 ** 2) + (13 ** 2)) ** 0.5**

이썬의 내장 함수인 `**` 연산자를 사용하여 제곱근을 계산할 수 있습니다. 

`**` 연산자는 왼쪽 피연산자를 오른쪽 피연산자만큼 거듭제곱하는 연산자입니다. 
제곱근은 0.5승을 의미하므로, `** 0.5`를 사용하여 제곱근을 계산할 수 있습니다.

따라서 밑변이 9cm, 높이가 13cm인 직각삼각형의 빗변을 구하는 코드는 다음과 같이 작성할 수 있습니다

```python
base = 9
height = 13

hypotenuse = ((base ** 2) + (height ** 2)) ** 0.5
```

위 코드에서 `(base ** 2 + height ** 2) ** 0.5`는 밑변의 제곱과 높이의 제곱을 더한 후, 그 결과에 0.5를 거듭제곱하여 제곱근을 계산하는 것입니다. 따라서 `hypotenuse`에는 빗변의 길이가 저장됩니다.

</aside>

### 3. 문제

**쉬지도 않고 일하는 직장인 오즈씨는 3년만에 첫 성과급으로 2500만원을 받았습니다. 연이율 6.15%의 고정금리 상품에 예금하려고 합니다. 3년 동안 넣어두면 원금을 제외한 이자가 얼마나 되는지 계산해 주세요**

(이자 지급 방식은 매년 원금 기준으로 지급되는 단리 방식입니다.)

원금 25000000원, 연이율 6.15%, 기간 : 3년

<aside>
🔑 정답 : **25000000 * 0.0615 * 3**

**이자를 계산하는 공식은 다음과 같습니다: `원금 * 이율 * 기간`입니다. 
이때 이율은 %로 계산을 위해서는 소수로 변환해야 합니다.** 

**6.15%는 0.0615로 변환하고 기간은 연 단위이므로 그대로 사용합니다.**

**예제 코드는 다음과 같습니다**

```python
principal = 25000000  # 원금 (25,000,000원)
interest_rate = 0.0615  # 연이율 (6.15%)
years = 3  # 기간 (3년)

interest = principal * interest_rate * years  # 이자 계산
```

</aside>
