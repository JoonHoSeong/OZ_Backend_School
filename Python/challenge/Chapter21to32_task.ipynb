### [프로젝트: 간단한 주소록 관리 시스템]

[목표]
- 사용자로부터 연락처 정보를 입력받아 관리하고,<br>
원하는 조건에 따라 연락처를 검색할 수 있는 간단한 주소록 관리 시스템을 만듭니다.

[기능]
- 연락처 추가: 사용자의 이름, 전화번호, 이메일 주소를 입력받아 주소록에 추가합니다.
- 연락처 목록 보기: 저장된 모든 연락처의 정보를 화면에 출력합니다.
- 연락처 검색: 사용자로부터 검색 키워드를 입력받아, 해당 키워드가 이름에 포함된 연락처 정보를 화면에 출력합니다.
- 연락처 삭제: 사용자로부터 이름을 입력받아, 해당하는 연락처 정보를 주소록에서 삭제합니다.

[구현 상세]
- 주소록 데이터는 리스트 내에 딕셔너리 형태로 저장됩니다.
- 각 기능은 별도의 함수로 구현합니다.
- 함수의 기본 활용, 가변 매개변수, 기본 매개변수, 키워드 매개변수, 타입 어노테이션,<br>
  제너레이터, 람다, 튜플, 조기 리턴, 재귀 함수, 함수 리턴 개념을 적절히 활용해야 합니다.

```python
# 가이드 코드입니다. 빈공간에 코드를 작성해주세요
address_book:list[dict[str, str, str]] = []

# 연락처 추가 함수
def add_contact(name : str, phone:str, email:str) -> None:
    address_book.append({'name' : name, 'phone' : phone, 'email':email})

# 연락처 목록 보기 함수
def view_contacts() -> None:
    for row in address_book :
        for col in row.keys() :
            print(f'{col} : {row[col]}')
        print('---------------')

# 연락처 검색 함수
def search_contacts(name :str) -> None:
    num = 0
    for row in address_book :
        if name.upper() in row['name'].upper() :
            num += 1
            for col in row.keys() :
                print(f'{col} : {row[col]}')
    if num == 0 :
        print('해당 키워드가 포함된 사람이 없습니다.')
    

# 연락처 삭제 함수
def delete_contact(name : str) -> None:
    for i,row in enumerate(address_book) :
        if name.upper() in row['name'].upper() :
            print(f"{row['name']}의 연착처를 삭제합니다.")
            print('---------------')
            del address_book[i]
    

# 메인 프로그램 루프
def main():
    while True :
        do_list = ['추가','보기','검색','삭제', '종료']
        do = input("원하는 기능을 입력해주세요(추가/보기/검색/삭제/종료)")
        if do not in do_list :
            continue
        #종료
        if do == do_list[-1] :
            print('프로그램을 종료합니다.')
            print('---------------')
            return 0
        #추가
        elif do == do_list[0] :
            print('주소록에 정보를 추가합니다.')
            name = input('이름을 입력해 주세요. : ')
            phone = input('전화번호를 입력해 주세요. :')
            email = input('이메일 주소를 입력해 주세요. :')
            print('---------------')
            add_contact(name=name, phone=phone,email=email)
        #보기
        elif do == do_list[1] :
            print("저장되어 있는 주소록을 출력합니다.")
            print('---------------')
            view_contacts()
        #검색
        elif do == do_list[2] :
            print("주소록의 정보를 검색합니다.")
            name = input('검색할 사람의 이름을 입력하세요. : ')
            print('---------------')
            search_contacts(name=name)
        #삭제
        else :
            print("주소록의 정보를 삭제합니다.")
            name = input('삭제할 사람의 이름을 입력하세요. : ')
            print('---------------')
            delete_contact(name=name)
        

if __name__ == "__main__":
    main()
```
