return 키워드 : 함수를 실행한 위치로 되돌아 감, 일반적으로 함수가 끝나는 위치에 return을 사용함

```python
# 값 없이 리턴하는 경우

def oz_return():
    print("파이썬")
    print("재미있엉")
    return
    print("거짓말하네")

oz_return()
```

1. oz_return()로 함수를 호출하면 함수안 코드가 위에서부터 하나씩 실행됨
2. 세번째에서 return을 만나 oz_return()으로 함수를 호출했던 그 위치로 돌아감

- 값과 함께 리턴하는 경우

```python
#값과 함께 리턴하는 경우
def oz_return():
    return "파이썬 재미있엉"

result = oz_return()
print(result)
```

- 아무것도 리턴하지 않는 경우

```python
def oz_return():
    return

result = oz_return()
print(result)
```

None : 아무것도 없다, bool()함수를 이용할 경우 False로 간주됨

```python
bool(None)
```
