시퀀스 자료형
 * 문자열 :   
        - 시퀀스형에서 사용할 수 있는 메소드 (count(), index(), split() 등...)<br>
        - 슬라이싱([ : ], [ : : ])<br>
        - 요소들로 이루어져 가능했던 기능들(요소 합치기, 요소 찾기 등등)<br> 

 * list
        - 시퀀스형에서 사용할 수 있는 메소드 (count(), index(), split() 등...)<br>
        - 슬라이싱([ : ], [ : : ])<br>
        - 요소들로 이루어져 가능했던 기능들(요소 합치기, 요소 찾기 등등)<br>
        - list 지원 메소드(append(), insert(), del, sort(), reverse())<br>
        - 2차원 배열( [[3,1], [2,1]])<br>

 * tuple
        - 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제 할 수 없다<br>
        - 값을 추가, 삭제, 변경할 수 없기 때문에 사용못하는 메소드들이 있다<br>
        - tuple의 생성방법은 tuple = 값, 값 또는 tuple = (값, 값)<br>
        - 그리고 엄청난 비밀을 숨기고 있다<br>
        - 튜플은 소괄호로 생성한다<br>

 * dict
        - 자료형의 정점에 있는 dict : Dictionary<br>
        - list, tuple, str, range(범위안에 연속된 데이터를 만드는 시퀀스 였어요)<br>
          공통적 특징이 있습니다. 데이터들의 연관성이 대부분 없었습니다.<br>
        - dict는 여러개의 값이 일렬로 정렬되면서도 값까리의 연관 관계를 고려할때 주로 사용<br>
          (예:  게임 케릭터 =>  힘:30, 지능:20, 체력:40, 민첩:90, 운:1) <br>
          => 케릭터라는 카테고리가 있죠<br>
        - dict 생성 방법은 중괄호 입니다.
        
* set
        - set은 수학의 집합을 의미합니다.<br>
        - 딕셔너리와 같이 순서가 없습니다.<br>
        - 값의 중복을 허락하지 않습니다. 아주 중요합니다.<br>
        - set 생성 방법도 중괄호를 사용합니다.<br>


```python
채소 = {"당근", "배추", "대파", "양파"}
type(채소)
print(채소)

# 순서가 없고, 출력하면 요소가 다르게 나오는것도 특징입니다.

#세트를 만드는 다양한 방법
세트 = set()
세트
세트2 = {}
세트2
```

```python
#중복을 제일 싫어하는 set

사과_리스트 = list("apple")
print(사과_리스트)
사과_튜플 = tuple("apple")
print(사과_튜플)
사과_set = set("apple")
print(사과_set)
```

```python
# set = 집합 , 합집합, 교집합, 차지합 기타 등등 ~!
# 대표적으로 합집합, 교집합, 차지합

채소1 = {"당근", "양파", "오이", "배추"}
채소2 = {"양파", "오이", "대파", "가자"}

#합집합(union)
채소1 | 채소2
set.union(채소1, 채소2)

#교집합(intersection)
채소1 & 채소2
set.intersection(채소1, 채소2)

#차집합(difference)
채소1 - 채소2
set.difference(채소1, 채소2)
```
