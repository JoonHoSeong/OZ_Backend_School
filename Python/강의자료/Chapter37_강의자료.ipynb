[math 모듈]

- 수학과 관련된 기능을 가지고 있음

math 모듈 내 함수
- sin(x) : 사인 값을 구함
- cos(x) : 코사인 값을 구함
- tan(x) : 탄젠트 값을 구함
- ceil(x) : 올림
- floor(x) : 내림

이 외에도 굉장히 많음....

````python
import math

print(math.sin(5))
print(math.cos(5))
print(math.ceil(3.4))
print(math.floor(5.2))
````

[from 구문]

모듈안에 수 많은 함수 중 우리가 원하는 함수만을 가져와 사용하면 문법적으로 굉장히 편해집니다.

```
from 모듈 이름 import 가져오고 싶은 변수 또는 함수
```

----
from math import sin

(기존)
math.sin(5) 

(from 구문 사용 후)
sin(5)

````python
from math import sin, cos, ceil, floor

print(sin(5))
print(cos(5))
print(ceil(3.4))
print(floor(5.2))
````

````python
#math 모듈의 모든 함수를 가져오는 방법으로 import에 '*' 사용합니다.
from math import *

print(sin(5))
print(cos(5))
print(ceil(3.4))
print(floor(5.2))
````

[as 구문]

모듈을 가져오다 보면 이름의 충돌이 발생하는 경우가 있습니다. 또는 너무 길어 짧게 줄여서 사용하고 싶은 경우도 있을겁니다.

이럴때 모듈의 식별자를 만들어 사용할 수 있는 문법이 as 구문입니다.

````python
import math as m

print(m.sin(5))
print(m.cos(5))
print(m.ceil(3.4))
print(m.floor(5.2))
````

[datetime 모듈]
- 날짜와 시간 관련된 모듈

````python
import datetime

print("오늘 날짜와 시간 출력")
now = datetime.datetime.now()
print(now.year, "년")
print(now.month, "월")
print(now.day, "일")
print(now.hour, "시")
print(now.minute, "분")
print(now.second, "초")
print()
````

````python
import datetime

print("다양한 시간 포맷")
output_1 = now.strftime("%Y.%m.%d %H:%M:%S")
output_2 = f'{now.year}년 {now.month}월 {now.day}일 {now.hour}시 {now.minute}분 {now.second}초'
output_3 = now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초")
print(output_1)
print(output_2)
print(output_3)
print()
````

[time 모듈]
- 시간과 관련된 기능을 다루는 모듈

````python
import time

print("3초 뒤에 어떤일이 일어날까요?")
time.sleep(3)
print("아무일도 없었다")
````
