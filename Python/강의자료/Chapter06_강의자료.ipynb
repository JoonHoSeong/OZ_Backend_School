메소드:어떤 명령을 수행하기 위해 작성된 코드들의 집합

쉽게 말해 붕어빵 메소드가 있다고 가정해보겠습니다.

붕어빵을 만들기 위해서는 반죽 넣기, 팥 넣기, 굽기와 같은 것들이 필요합니다.

반죽 넣기 코드, 팥 넣기 코드, 굽는 코드 등을 작성해서 하나로 묶어
놓으면 붕어빵 메소드라는 것만 실행하면 자동적으로 안에 있는 기능들이 작동하면서 붕어빵이 만들어지겠죠?

이렇게 매번 붕어빵을 만들기 위해 반죽 넣기 코드, 팥 넣기 코드, 굽는 코드 등을 만들지 않도록 미리 만들어 놓는것이 메소드 입니다.

```python
숫자 = "98852116"
#전체 출력
print(숫자[::])

#전체 역순으로 출력하기
print(숫자[::-1])

#첫번째 인덱스 기준으로 2칸씩 띄어서 출력하기
print(숫자[::2])

#4번째 인덱스에서 1번 인덱스까지 거꾸로 출력
print(숫자[4:1:-1])
```
