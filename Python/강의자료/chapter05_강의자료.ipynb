### 변수에 값을 넣는 것 => 할당

sum = 1 + 2는 1+2의 계산된 결과를 sum이라는 변수에 할당한다고 설명할 수 있습니다.
sum 변수명이고 "=" 같다는 의미가 아니라 할당(assignment)이라고 합니다.  

- "=" : 할당 연산자  
- "+" : 산술 연사자  

+= 산술 후 바로 할당하는 문법  
/=, -=, *= 모두 가능합니다.

```python
# 변수를 사용하면서 사용 가능한 문법

sum = 1 + 2
sum = sum + 3
sum = sum + 4
sum = sum + 5

print(sum)

sum2 = 1 + 2
sum2 += 3
sum2 += 4
sum2 += 5

print(sum2)
```

```python
# 주의 !! 선언되지 않은 변수를 사용할 경우 에러 발생

sum3 = 0
sum3 += 1

sum3
```
