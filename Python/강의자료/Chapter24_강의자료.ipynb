```python
#(기본 매개변수, 가변 매개변수)

def oz(n=3, *values):
    for i in range(n):
        for value in values:
            print(value)
        print()

oz("기본 매개변수 다음에", "가변 매개변수를", "넣을 수 있을까요?")
```

오류 메시지는 "숫자형 데이터가 들어와야하는데 문자형 데이터가 들어와 오류가 발생"이라고 해석 할 수 있습니다.

이 오류가 발생했다는건 oz("기본 매개변수 다음에", "가변 매개변수를", "넣을 수 있을까요?")로<br>
함수를 호출했을때 <span style='color:red'>n에 "기본 매개변수 다음에"값이 들어가고</span><br><span style='color:yellow'>*values에 "가변 매개변수를", "넣을 수 있을까요?"가</span> 들어 갔기 때문입니다.

--- 
정리해보자면 각 식별자에는 이러한 데이터가 들어가게 됩니다.

oz("기본 매개변수 다음에", "가변 매개변수를", "넣을 수 있을까요?")<br>
<span style="font-weight:bold">n</span> = "기본 매개변수 다음에"<br>
<span style="font-weight:bold">*values</span> = "가변 매개변수를", "넣을 수 있을까요?"

결국 기본 매개변수가 가변 매개변수보다 앞에 오게 되면 기본 매개변수는 의미가 없어집니다.


그렇다면, 반대로 기본 매개변수가 가변 매개변수보다 앞에 오게 되면 어떻게 될까요?

oz함수에 3개의 가변 매개변수와 하나의 기본 변수로 이루어진 값을 넣어보겠습니다.

입력값 : oz("기본 매개변수 다음에", "가변 매개변수를", "넣을 수 있을까요?", 10)

결과는 어떻게 될까요?

```python
#(가변 매개변수, 기본 매개변수)
def oz(*values, n=3):
    for i in range(n):
        for value in values:
            print(value)
        print()

oz("기본 매개변수 다음에", "가변 매개변수를", "넣을 수 있을까요?", 10)
```

기본 매개변수에 10을 넣으려 했으나 결과는 3번만 출력되었습니다.

oz() 함수 호출시 입력된 매개변수들이 모두 가변 매개변수로 인식되어 oz() 함수 정의시 입력된 기본 매개변수 3이 적용되었습니다.

이러한 문제를 해결하기 위해서 나온 개념이 <span style="color:red">키워드 매개변수</span>입니다.

<br><br>
앞서 몇번 언급했던 print()함수를 다시 한번 보면 기본 매개변수라 꽤 있습니다.<br>
그 중에서도 end는 익숙합니다.<br>


```python
#end = \n
list = ["안", "녕", "하", "세", "요"]
for i in list:
    print(i)
```

```python
list = ["안", "녕", "하", "세", "요"]
for i in list:
    print(i, end="")
```

print() 함수를 호출할때 i는 가변 매개변수로 입력되고 뒤에 나오는 end="" 기본 매개변수가 됩니다.<br>
어떻게 가변 매개변수 뒤에 기본 매개변수가 될 수 있었을까요?

앞서 예시 코드와 다른점은 end=""라는 형식으로 값을 지정한 부분입니다.<br>
end="" 이 부분이 오늘 수업의 핵심인 <span style="color:red">키워드 매개변수</span>입니다.

```python
#직육면체를 부피를 구하는 코드를 작성해보자 : 가로 : x, 세로 : y, 높이 : h
def oz(x, y=20, h=10):
    print(x * y * h)

#x식별자에 매개변수 10을 넣은 경우
oz(10)   
#x식별자에 10을 넣고 y=10이라는 키워드 매개변수를 이용한 경우
oz(10, y=10)
#모든 매개변수를 키워드 매개변수로 넣은 경우
oz(x=5, y=5, h=5)
```

키워드 매개변수는 이렇게 다양하게 사용될 수 있습니다.<br>
