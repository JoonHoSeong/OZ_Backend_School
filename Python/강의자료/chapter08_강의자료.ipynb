```python
#본격적인 논리연사자
# True : 1 / False : 0
True and True # 1 * 1 = 1
True and False # 1 * 0 = 0
False and True # 0 * 1 = 0
False and False # 0 * 0 = 0
```

```python
#본격적인 논리연산자
# True : 1 / False : 0
True or True # 1 + 1 = 1
True or False # 1 + 0 = 1
False or True # 0 + 1 = 1
False or False # 0 + 0 = 0
```
