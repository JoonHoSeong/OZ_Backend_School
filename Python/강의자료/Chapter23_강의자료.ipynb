*values 뒤에 오는 매개변수인 <span style="color : red"> sep:str | None = " ", end: str | None = "\n" </span>들은<br>
마치 값을 할당하는 형태 처럼 보이는 이 매개변수들은 모두 <span style="font-weight:bold">기본 매개변수</span>입니다.

기본 매개변수는 매개변수를 입력하지 않으면 함수 정의시 설정해 놓은 기본값으로 대체됩니다.<br>
역시나, 기본 매개변수 사용시에도 제약이 있습니다.
- 기본 매개변수 뒤에는 일반 매개변수가 올 수 없다.

```python
def oz(value, n=2):
    for i in range(n):
        print(value)

oz("일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.")
```

```python
def oz(value, n=2):
    for i in range(n):
        print(value)

oz("일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.", n=10)
```

oz() 함수에서 기본 매개변수를 사용했기 때문에 value 식별자에 넣을 값만 입력 후 oz() 함수를 호출해도<br>
오류가 발생하지 않고 정상적으로 작동하는것을 확인할 수 있습니다.

그럼 기본 매개변수 뒤에 일반 매개변수를 배치해서 실행하면 어떻게 될까요?

```python
def oz(n=2, value):
    for i in range(n):
        print(value)

oz("일반 매개변수 다음에 기본 매개변수를 배치해야 합니다.")
```
