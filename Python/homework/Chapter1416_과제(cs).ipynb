[문제2]

Python의 얕은 복사와 깊은 복사에 대해서 설명하세요


파이썬의 대입 문은 객체를 복사하는 것이 아니라 대상과 객체 사이에 바인딩을 생성합니다.
불변객체에는 얕은 복사와 깊은 복사에 차이가 없지만 가변객체의 경우 차이가 있습니다.
얕은 복사는 같은 객체 값을 바인딩만 하는 역할을 합니다.
깊은 복사는 새로운 객체를 생성하고 원복에 있는 값들을 복사하는 것을 의미합니다.
이러한 깊은 복사의 특징으로 같은 값을 가진 두변수가 생기지만 가리키는 객체가 달라 독립적으로 사용할 수 있습니다.
