### 1. Python의 변수 할당 개념을 다른 언어와 비교해 설명해주세요

<aside>
🔑 정답 :

파이썬의 변수 할당 개념은 다른 프로그래밍 언어와 비교했을 때 몇 가지 특징이 있습니다.

파이썬에서 변수는 특정 값을 참조하는 이름으로써 동작합니다. 

이는 일부 다른 언어(예: C 언어)에서 변수가 메모리의 특정 영역을 가리키는 것과는 다릅니다. 
파이썬에서 변수를 할당하면 해당 변수는 오른쪽에 있는 표현식의 결과를 참조하게 됩니다.

예를 들어, `x = 5`라고 하면 `x`는 값 `5`를 참조하게 됩니다. 그리고 나서 `y = x`라고 하면 `y`는 `x`가 참조하는 것과 같은 `5`를 참조하게 됩니다. 이후 `x`의 값을 변경해도 `y`의 값은 변하지 않습니다. 이는 `x`와 `y`가 독립적으로 같은 값을 참조하고 있기 때문입니다.

</aside>

### 2. Python의 실수형(float) 계산 시 오차가 발생하는 이유에 대해 간단히 설명해 주세요

<aside>
🔑 정답 :

파이썬을 포함한 대부분의 컴퓨터 프로그래밍 언어에서 실수형(`float`) 계산 시 오차가 발생하는 주된 이유는 컴퓨터가 실수를 이진수(2진수) 형태로 표현하고 저장하기 때문입니다.

컴퓨터는 모든 데이터를 이진수로 변환하여 저장하고 처리합니다. 그러나 모든 실수를 정확하게 이진수로 표현하는 것은 불가능합니다. 

특히, 10진수 체계에서는 간단하게 표현되는 실수가 이진수에서는 무한히 반복되는 분수로 표현될 수 있습니다. 예를 들어, 10진수에서는 0.1이라는 값이 간단하지만, 이진수에서는 이 값을 정확하게 표현하기 위해 무한히 반복되는 소수가 필요합니다.

따라서 컴퓨터는 이런 값들을 가장 근사한 이진수로 반올림하여 저장하게 되는데, 이 과정에서 소수점 이하의 매우 작은 오차가 발생하게 됩니다. 이 오차는 대부분의 경우 문제가 되지 않지만, 많은 수의 실수를 계산하는 과정에서 이 오차가 누적되면 결국 큰 오차로 이어질 수 있습니다.

</aside>

### 3. 슬라이싱에 대해 간단히 설명해주세요

<aside>ㅇ
🔑 정답 :

파이썬의 슬라이싱은 시퀀스 타입의 일부분을 선택하거나 추출하는 기능을 말합니다. 시퀀스 타입에는 리스트, 문자열, 튜플 등이 있습니다.

슬라이싱의 기본 형태는 `시퀀스[시작:끝:간격]`입니다.

슬라이싱을 이용하면 우리는 시퀀스의 일부분을 쉽게 선택하거나 추출할 수 있습니다. 이는 데이터 처리나 문자열 처리 등 다양한 상황에서 유용하게 사용됩니다.

</aside>
