### 1. **list와 tuple의 주된 차이점 은 무엇인가요?**

<aside>
🔑 정답 :

리스트와 튜플의 주된 차이점은 변경 가능성, 즉 '가변성(mutability)'입니다.

리스트는 가변적입니다. 
리스트의 요소는 생성된 후에도 추가, 수정, 삭제가 가능합니다. 예를 들어, `list1 = [1, 2, 3]`이라는 리스트가 있을 때, `list1[1] = 5`와 같은 방식으로 리스트의 요소를 변경하거나 `list1.append(4)`와 같은 방식으로 리스트에 요소를 추가할 수 있습니다. 

반면 튜플은 불변적입니다. 튜플의 요소는 생성된 후에 변경할 수 없습니다.
예를 들어, `tuple1 = (1, 2, 3)`이라는 튜플이 있을 때, 이 튜플의 요소를 변경하려고 하면 TypeError가 발생합니다.

</aside>

### 2. **Python의 set에 대해 설명해 주세요**

<aside>
🔑 정답 :

세트(Set)는 중복된 값을 가질 수 없는 자료형입니다. 
세트는 항상 고유한 요소만을 저장하며 순서가 없습니다. 따라서 세트의 요소는 인덱싱을 통해 접근할 수 없습니다.

세트는 중괄호 `{}`를 사용하여 생성하거나 `set()` 함수를 사용할 수 있습니다. 
예를 들어 `my_set = {1, 2, 3}` 또는 `my_set = set([1, 2, 3])`으로 세트를 생성할 수 있습니다.

또한, 세트는 수학의 집합 연산(합집합, 교집합, 차집합 등)을 수행하는 데 유용하게 사용됩니다. 
예를 들어, 두 세트 A와 B가 있을 때, `A.union(B)` 또는 `A | B`는 A와 B의 합집합을, `A.intersection(B)` 또는 `A & B`는 A와 B의 교집합을, `A.difference(B)` 또는 `A - B`는 A에서 B를 뺀 차집합을 반환합니다.

</aside>

### 3. **Python의 list는 연결리스트(linked list) 인가요?**

<aside>
🔑 정답 :

파이썬의 리스트는 연결 리스트(linked list)가 아닙니다. 
파이썬의 리스트는 내부적으로 동적 배열(dynamic array)을 사용하여 구현되어 있습니다.

동적 배열은 미리 일정량의 메모리를 할당받아 사용하고, 필요에 따라 크기를 조절하는 데이터 구조입니다. 파이썬의 리스트는 이 동적 배열을 통해 요소의 추가, 삭제, 수정 등의 연산을 수행합니다.

반면에 연결 리스트는 각 요소가 다음 요소를 가리키는 방식으로 구현되는 자료구조로, 요소의 추가와 삭제가 자유롭지만, 특정 요소에 접근하려면 처음부터 순차적으로 탐색해야 하는 단점이 있습니다.

따라서, 파이썬의 리스트는 연결 리스트와는 다른 동적 배열을 기반으로 하는 자료구조입니다.

</aside>

### 4. **딕셔너리 자료형에 대해 설명해주세요**

<aside>
🔑 정답 : 
****

**파이썬의 딕셔너리(Dictionary) 자료형은 키(key)와 값(value)의 쌍을 요소로 가지는 자료형입니다. 
딕셔너리는 해시 테이블(Hash Table)로 구현되어 있어, 키를 통해 값을 빠르게 검색할 수 있는 장점이 있습니다.**

**딕셔너리는 중괄호 `{}` 안에 `키: 값` 형태로 요소를 작성하여 생성합니다. 예를 들어, `my_dict = {'oz': 1, 'oz_FE': 2, 'oz_BE': 3}`와 같이 딕셔너리를 생성할 수 있습니다.**

**딕셔너리의 키는 변경할 수 없으며, 딕셔너리 내에서 고유해야 합니다. 
즉, 하나의 딕셔너리 내에 동일한 키가 두 개 이상 존재할 수 없습니다. 반면에 값은 변경 가능하며, 동일한 값이 여러 개 존재할 수 있습니다.**

**딕셔너리의 요소에 접근하려면 `딕셔너리[키]` 형태로 접근하며, 특정 키의 값을 변경하려면 `딕셔너리[키] = 값` 형태로 변경할 수 있습니다. 
또한 `del 딕셔너리[키]` 형태로 특정 키를 가진 요소를 삭제할 수 있습니다.**

</aside>
