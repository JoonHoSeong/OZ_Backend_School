### 1. 파이썬 2.7 버전과 파이썬 3에서 range 차이를 설명해주세요

<aside>
🔑 정답 : 

****파이썬 2.7에서 `range` 함수는 지정된 범위의 숫자 목록을 메모리에 생성합니다. 이는 큰 범위의 숫자를 생성할 때 메모리 문제를 일으킬 수 있습니다. 예를 들어, `range(1, 10000000)`은 1부터 9999999까지의 모든 숫자를 메모리에 저장하려고 시도합니다.

반면에 파이썬 3에서 `range` 함수는 "range" 객체를 반환합니다. 이 객체는 숫자를 필요에 따라 동적으로 생성하므로 메모리를 효율적으로 사용합니다. 즉, 파이썬 3의 `range` 함수는 숫자를 미리 생성하지 않고, 숫자가 필요할 때마다 생성하여 반환합니다. 이런 방식을 'lazy evaluation'이라고 합니다.

따라서 파이썬 2.7에서 큰 범위의 숫자를 생성할 때는 `xrange` 함수를 사용하는 것이 더 효율적일 수 있습니다. `xrange` 함수는 파이썬 3의 `range` 함수와 유사하게 작동하며, 숫자를 필요에 따라 동적으로 생성합니다. 하지만 파이썬 3에서는 `xrange` 함수가 없고, 대신 `range` 함수가 `xrange` 함수의 기능을 포함하고 있습니다.

</aside>

### 2. 얕은 복사와 깊은 복사에 대해서 설명하세요

<aside>
🔑 정답 :

얕은 복사 (Shallow Copy):
얕은 복사는 원본 객체의 '참조'를 복사하는 것을 말합니다. 이 경우, 복사된 객체와 원본 객체는 같은 메모리 주소를 공유하게 됩니다. 따라서 복사된 객체에서 변경이 이루어지면 원본 객체에도 그 변경사항이 반영됩니다. 파이썬에서는 `copy` 모듈의 `copy()` 함수나 리스트의 `[:]` 연산을 통해 얕은 복사를 수행할 수 있습니다.

깊은 복사 (Deep Copy):
깊은 복사는 원본 객체의 '값'을 복사하는 것을 말합니다. 이 경우, 복사된 객체와 원본 객체는 서로 다른 메모리 주소를 가집니다. 따라서 복사된 객체에서 변경이 이루어져도 원본 객체에는 영향을 미치지 않습니다. 파이썬에서는 `copy` 모듈의 `deepcopy()` 함수를 통해 깊은 복사를 수행할 수 있습니다.

즉, 얕은 복사는 원본과 복사본이 서로 영향을 주고 받는 반면, 깊은 복사는 원본과 복사본이 서로 독립적인 관계를 유지합니다.

</aside>
