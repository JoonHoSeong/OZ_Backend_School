[문제3]

클래스 상속에 대해 설명해주세요

클래스의 상속은 부모클래스의 내용을 자식 클래스가 가지게 되는것이다.
자식클래스를 선언할때 소괄호로 부모클래스를 포함시키면 상속 된다.
상속을 받으면 자식 클래스는 부모 클래스에 있는 속성과 메소드는 쓰지 않아도 사용이 가능하다.

[문제4]

클래스 다중 상속과 관련된 MRO에 대해 설명해주세요

클래스 다중 상속
하나의 클래스가 두개 이상의 부모 클래스로 부터 상속을 받는 것을 의미한다.

MRO
자식 클래스가 다중상속을 받은뒤  부모클래스의 동일한 이름을 가진 메소드를 사용하려 할때 동일한 이름을 가진 각각의 부모클래스내의 메소드중 어느것을 써야할지 모를때 발생하는 문제를 해결하기 위한 것이다.
MRO는 자식과 부모 클래스를 전부 포함하여 메소드의 실행 순서를 지정한다.
